```python
import pandas as pd
import joblib
```

```python
# Load saved model, vectorizer, and label encoder
best_model = joblib.load("saved_model_fin.pkl")
tfidf_vectorizer = joblib.load("tfidf_vectorizer.pkl")
label_encoder = joblib.load("label_encoder.pkl")
```

```python
# Load raw data
raw_data = pd.read_excel("banklast_quarter.xlsx")
```

```python
#Cleaning the data
raw_data = raw_data.dropna(subset=["TRANSACTION DETAILS"])  # Ensure no missing transaction details
X_raw_text = raw_data["TRANSACTION DETAILS"]

# Transform data using TF-IDF vectorizer
X_raw = tfidf_vectorizer.transform(X_raw_text)

# Predict tags
predictions = best_model.predict(X_raw)

# Decode predicted labels
predicted_tags = label_encoder.inverse_transform(predictions)

# Save predictions to a CSV file
raw_data["Predicted Category"] = predicted_tags
```

```python
# Save predictions
raw_data.to_csv("classified_transactions.csv", index=False)
print("Predictions saved to classified_transactions.csv")
```
